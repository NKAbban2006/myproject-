```python
plt.figure(figsize=(10,6))
plt.hist(redwine['alcohol'], bins=20, color='skyblue')
plt.title('Distribution of Alcohol content')
plt.xlabel('Alcohol content (%)')
plt.ylabel('Frequency')
plt.show()           
```

```python
x = redwine['alcohol']
y = redwine['chlorides']
plt.xlabel('alcohol(%)', fontsize=18)
plt.ylabel('chlorides(out of 1)', fontsize=16)
plt.title('A chart of alcohol percentage against chloride content')
plt.bar(x, y)
```

# __OBSERVATION__
### The histogram above displays the relative comparison between several samples with respect to alcohol percentage. It can be observed from the histogram that majority of the samples contain roughly between 9 and 10% alcohol content. The bar chart shows the relative comparisons of the chloride contents of various samples. It can be observed that, most of the samples have a chloride content of roughly 0.6.
